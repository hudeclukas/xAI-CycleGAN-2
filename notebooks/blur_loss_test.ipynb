```python
import cv2
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
img = cv2.imread('img.png', 0)
plt.imshow(img, cmap='gray')
```

```python
cv2.Laplacian(img, cv2.CV_64F).var()
```

```python
img_blur = cv2.blur(img, (5, 5))
plt.imshow(img_blur, cmap='gray')
```

```python
cv2.Laplacian(img_blur, cv2.CV_64F).var()
```

```python
he = cv2.imread('he.png', 0)
he_blur = cv2.imread('he_blur.png', 0)
```

```python
plt.imshow(he, cmap='gray')
```

```python
plt.imshow(he_blur, cmap='gray')
```

```python
cv2.Laplacian(he, cv2.CV_64F).var()
```

```python
cv2.Laplacian(he_blur, cv2.CV_64F).var()
```

```python
# I guess this is a good way to produce a 'bluriness' loss
max(0, cv2.Laplacian(he, 3).var() / cv2.Laplacian(he_blur, 3).var() - 1)
```
